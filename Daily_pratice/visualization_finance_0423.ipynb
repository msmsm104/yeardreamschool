## Matplotlib
> - 파이썬의 대표적인 시각화 패키지



```python
# 파이썬 라이브러리 3대장
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### line plot
    line plot은 시간 혹은 순서에 따라 데이터가 어떻게 변화하는지를 보기위해 주로 사용합니다. 
    ex)주가, 전력수요량, 매출, 수요


```python
# 2차함수를 그려보자.

x = np.linspace(-10, 10, 100)
y = x ** 2
```

```python
plt.figure(figsize=(8, 6))

plt.plot(x, y)

plt.show()
```

### 스타일 지정

```python
plt.figure(figsize=(8, 6))

plt.plot(x, y, 'r--')

plt.show()
```

### 그래프에 다양한 옵션 추가하기

```python
plt.figure(figsize=(12, 6)) # 그래프의 크기조정

plt.plot(x, y, 'r--') # 그래프 설정

plt.show() # 그리기 마무리
```

```python
# 타이틀
plt.figure(figsize=(8, 6))

plt.title('$x^2$')
plt.plot(x, y, 'r--')

plt.show()
```

```python
# 축 타이틀 지정

plt.figure(figsize=(8, 6))

plt.title('$x^2$')
plt.plot(x, y, 'r--')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
# 범례지정
# x^3그래프를 같은곳에 그려보자. - 제대로된 그래프의 모양을 관측하기 힘들다.

y1 = x ** 3


plt.figure(figsize=(8, 6))

plt.title('$x^2$')
plt.plot(x, y, 'r--', label='$x^2$')
plt.plot(x, y1, '--', label='$x^3$')

plt.legend(loc='best') # 범례표시

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
# 두가지 그래프의 모양이 제대로 관측되지 않으므로 축 범위를 설정하여 해결하자.

plt.figure(figsize=(8, 6))

plt.title('$x^2$')
plt.plot(x, y, 'r--', label='$x^2$')
plt.plot(x, y1, '--', label='$x^3$')

plt.legend(loc='best')

plt.xlim(-5, 5)
plt.ylim(-10, 10)

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
# 축 구간 지정
# 그리드 추가

plt.figure(figsize=(8, 6))

plt.title('$x^2$')
plt.plot(x, y, 'r--', label='$x^2$')
plt.plot(x, y1, '--', label='$x^3$')

plt.grid()

plt.legend(loc='best')

plt.xticks([-5,-2.5,0,2.5,5])
plt.yticks(range(-10,10,2))

plt.xlim(-5, 5)
plt.ylim(-10, 10)

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

### 연습문제
> y축에 x의 세제곱 그래프를 그려봅니다.  
> 라인은 노랑색, 선은 실선, 마커는 없어요.   
> 제목은 '$X^3$', plot label=''$x^3$' 범례지정도 해주세요

```python
x_1 = np.linspace(-10,10,100)
y_1 = x_1 ** 3
```

```python
plt.figure(figsize=(8,6))

plt.title('$X^3$')

plt.grid()

#x, y축의 범위도 추가적으로 지정해보자
plt.xlim([-5,5])
plt.ylim([-25, 25])

plt.plot(x_1, y_1, 'y', label='$x^3$')

plt.legend(loc='best')

plt.show()
```

### subplot 연습문제
> 2 x 2 figure subplot설정  
> 제곱, 세제곱, sin, cos 그래프 나누어 그려봅시다

```python
x_2 = np.linspace(-10, 10, 1000)
y_2 = x_2 ** 2
y_3 = x_2 ** 3
y_4 = np.sin(x_2)
y_5 = np.cos(x_2)


plt.figure(figsize=(16, 12))

plt.grid()

plt.subplot(2,2,1)
plt.title('$x^2$')

plt.xlim([-5, 5])
plt.ylim([-10,10])
plt.grid()
plt.plot(x_2, y_2)

plt.subplot(2,2,2)
plt.title('$x^3$')
plt.grid()
plt.plot(x_2, y_3)

plt.subplot(2,2,3)
plt.title('$sin(x)$')
plt.grid()
plt.plot(x_2, y_4)

plt.subplot(2,2,4)
plt.title('$cos(x)$')
plt.grid()
plt.plot(x_2, y_5)





plt.show()
```

### bar plot
> bar plot은 범주형 데이터(구간이 나누어진) 데이터 갯수, 빈도를 나타내는데 사용.
> loan 데이터의 신용등급을 bar plot으로 그려보자.

```python
# loan1.csv 파일 불러오기

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/이어드림스쿨/0422_인공지능 파이썬 advance/0422_파이썬 abvance_강의자료/advance/data/loan1.csv')
df.head()
```

```python
# 해당 데이터프레임의 범주형 데이터 세부신용등급(sub_grade)에 접근하여 어떤식으로 나누어져 있는지 확인해보자.

df['sub_grade'].head(10)

## 알파벳으로 각 신용등급이 나누어져 있음을 확인할 수 있다.
## 각 신용등급의 갯수를 파악해보자
```

```python
df['sub_grade'].value_counts()
```

```python
# 해당 수치형 데이터를 그래프(bar plot)로 나타내보자

grade_idx = df['sub_grade'].value_counts().sort_index().index
grade_val = df['sub_grade'].value_counts().sort_index().values

plt.figure(figsize=(8, 6))

plt.bar(grade_idx, grade_val)

plt.show()
```

### 히스토그램

```python
# 대출금액분포(loan_amnt)를 hist plot(히스토그램)으로 확인해보자.

plt.figure(figsize=(8, 6))

plt.hist(df['loan_amnt'], bins=25)

plt.show()
```

### scatter plot

> 2개 혹은 2개 이상의 변수간의 상관관계를 확인하고자 할 때 많이 사용하는 시각화방법.

```python
# 대출금액(loan_amnt)와 원리금: 원금과 이자를 합친돈(installment)의 관계를 scatter plot을 통해 알아보자
# 상식적으로 생각했을때 양의 상관관계를 가지고 있을것이라고 예측된다.

plt.figure(figsize=(8, 6))

plt.scatter(df['loan_amnt'], df['installment'])

plt.show()
```

> - 그래프에서 확인할 수 있듯이 두 변수를 양의 상관관계
> - 그래프상에 점 하나가 하나의 샘플(대출건수)이다.

## seaborn

> - matplotlib에서 지원하지 않는 다양한 형태의 시각화도구를 제공한다.

```python
import seaborn as sns
```

### barplot

```python
x = df['sub_grade'].value_counts().sort_index().index
y = df['sub_grade'].value_counts().sort_index().values

plt.figure(figsize=(12,6))

sns.barplot(x, y)

plt.show()
```

### countplot

```python
plt.figure(figsize=(12,6))

sns.countplot(df['grade'])

plt.show()
```

### displot(히스토그램)

```python
plt.figure(figsize=(12,6))

sns.displot(data=df, x='loan_amnt', kde=True)

plt.show()
```

### scatterplot

```python
plt.figure(figsize=(12,6))

sns.scatterplot(data=df, x='loan_amnt', y='installment', hue='grade', alpha=0.4)
## hue='grade' grade별로 색을 나눠서 보여줘.

plt.show()
```

### boxplot
> - box하나당 해당 데이터의 사분위수, 중앙값, 최대, 최소값, 아웃라이어 등을 표기한다.

```python
plt.figure(figsize=(12, 10))

sns.boxplot(data=df, x='grade', y='int_rate')

plt.show()
```

> - 등급별 이자율의 범위가 정해져있다.
> - 등급이 낮아도 이자율을 적게 내는 사람이 있다. (정상적인 데이터가 맞는 것일까?)

```python
# 이상치에 대한 샘플링

cond_1 = df['int_rate'] < 7
cond_2 = df['grade'] != 'A'

df[cond_1 & cond_2][['int_rate', 'emp_title']]
```

> - 이상치에 해당하는 샘플들의 이자율은 6%로 동일(고정금리)


### jointplot

> - loan_amnt, int_rate의 scatterplot+hist 인 jointplot을 그려보자.

```python
plt.figure(figsize=(12, 10))


sns.jointplot(data=df, x='loan_amnt', y='int_rate', kind='hex')
## kind = [kde, hist, hex, reg, resid]

plt.show()
```

### pairplot
> - 각 변수간 상관관계를 확인하는 scatterplot의 확장버젼


```python
sliced_df = df[df.columns[:25]]
```

```python
# 각 column간의 모든 scatter plot을 그린다.

plt.figure(figsize=(12, 10))

sns.pairplot(data=sliced_df)

plt.show()
```

### heatmap
> 각 변수간의 상관관계를 열지도로 표현

```python
plt.figure(figsize=(12, 10))

sns.heatmap(df.corr())

plt.show()
```
