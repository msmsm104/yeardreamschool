```python
v1 = np.array([1,2,3,4])
v2 = np.array([1,2,1,1])
```

```python
data = data.drop(columns='Id')
data.head()
```

```python
def euclidian_distance(p, q):
    return np.linalg.norm(p - q) #벡터의 크기 (L2 distance)

euclidian_distance(v1, v2)
```

```python
def cosine_similarity(p, q):
    # 1. 분자 
    dot = p @ q
    # 2. 분모
    norm = np.linalg.norm(p) * np.linalg.norm(q) + 1e-12
    return dot / norm

cosine_similarity(v1, v2)
```
